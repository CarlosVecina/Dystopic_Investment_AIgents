```python
import datetime
from src.data_ingestion.download_ticker_data import download_ticker_data

df = download_ticker_data("TSL", datetime.datetime(2024,1,1), datetime.datetime(2024,4,3))
```

```python
from src.data_ingestion.transform_basic_indicators import calculate_bollinger_bands, calculate_ema, calculate_macd, calculate_rsi, calculate_sma


df['SMA'] = calculate_sma(df['Close'])
df['EMA'] = calculate_ema(df['Close'])
upper_band, lower_band = calculate_bollinger_bands(df['Close'])
df['Upper Bollinger Band'] = upper_band
df['Lower Bollinger Band'] = lower_band
df['RSI'] = calculate_rsi(df['Close'])
macd_line, signal_line = calculate_macd(df['Close'])
df['MACD Line'] = macd_line
df['MACD Signal Line'] = signal_line
```
